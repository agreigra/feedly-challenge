```python
TAG_RE = re.compile(r'<[^>]+>')

def remove_tags(text):
    return TAG_RE.sub('', text)
```

```python
nltk.download('stopwords')
nltk.download('punkt')
GARBAGE = {"'s", "n't", '...', 'oh',"'m", "'re", "'", "''", "'ve", "'ll", "'d", "``" }
STOP_WORDS = set(stopwords.words('english')).union(GARBAGE)


def extract_tokens(text):
    res = []
    for sent in nltk.sent_tokenize(text):
        tmp_res = nltk.word_tokenize(sent)
        for token in tmp_res:
            res += re.split("[\./]", token)
    return res

def clean_tokens(tokens):
    return [token.lower() for token in tokens if token not in string.punctuation]

def remove_stop_words(tokens):
    return [token for token in tokens if token not in STOP_WORDS]

def text2tokens(text):
    tokens = extract_tokens(text)
    tokens = clean_tokens(tokens)
    tokens = remove_stop_words(tokens)
    return tokens
```

```python
df["content"] = df["content"].apply(lambda s: remove_tags(s))
```

```python
df["tokens"] = df["content"].apply(lambda s: text2tokens(s))
```

```python
words_to_id = dict()
for _, tokens in df["tokens"].iteritems():
    for word in set(tokens):
        if word not in words_to_id:
            words_to_id[word] = len(words_to_id)
```

```python
def bag_of_words_to_vector(bag_of_words, words_to_id):
    vec = np.zeros(len(words_to_id))
    counts = Counter(bag_of_words)
    for word, count in counts.items():
        if word in words_to_id:
            vec[words_to_id[word]] = count
    return vec
```

```python
def generate_features(df, words_to_id):
    features = []
    for _, row in df.iterrows():
        features.append(bag_of_words_to_vector(row["tokens"], words_to_id))
    return np.array(features)
```

```python
data_train, data_valid = sklearn.model_selection.train_test_split(df, test_size=0.2)
```

```python
X_train= generate_features(data_train, words_to_id)
y_train = data_train["label"]
```
